```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path="/content/spotify-dataset.csv"
df=pd.read_csv(file_path)
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df_cleaned=df.dropna()
```

```python
df_cleaned.isnull().sum()
```

```python
plt.figure(figsize=(8,4))
sns.countplot(x='explicit',data=df_cleaned, hue='explicit',palette='coolwarm')
plt.title('Explicit Label Statistics')
# plt.xlabel('Explicit')
plt.ylabel('No. of Songs')
plt.show()
```

```python
plt.figure(figsize=(8,4))
sns.histplot(df_cleaned['liveness'],bins=30,color='orange',kde=True)
plt.title('Distribution of Liveness')
plt.xlabel('Liveness')
plt.ylabel('No. of Songs')
plt.show()
```

```python
plt.figure(figsize=(8,4))
sns.histplot(df_cleaned['popularity'],bins=30,kde=True)
plt.title('Distribution of Popularity')
plt.xlabel('Popularity')
plt.ylabel('No. of Songs ')
plt.show()
```

```python
plt.figure(figsize=(8,4))
sns.histplot(df_cleaned['danceability'],bins=30,kde=True, color="Yellow")
plt.title('Distribution of danceability')
plt.xlabel('Danceability')
plt.ylabel('No. of Songs')
plt.show()
```

```python
plt.figure(figsize=(8,4))
sns.histplot(df_cleaned['energy'],bins=20,kde=True, color="Red")
plt.title('Distribution of Energy Levels Across Song ')
plt.xlabel('Energy')
plt.ylabel('No. of Songs')
plt.show()
```

```python
numeric_df=df_cleaned.select_dtypes(include=[np.number])
co_matrix=numeric_df.corr()
plt.figure(figsize=(12,10))
sns.heatmap(co_matrix,annot=True,cmap="viridis")
plt.title('Correlation Matrix')
plt.show()
```
